```python
from collections import defaultdict

# Creating a dictionary to represent the adjacency list of the graph
graph = defaultdict(list)
graph['Arad'] = ['Zerind', 'Sibiu', 'Timisoara']
graph['Zerind'] = ['Arad', 'Oradea']
graph['Oradea'] = ['Zerind', 'Sibiu']
graph['Sibiu'] = ['Arad', 'Oradea', 'Fagaras', 'Rimnicu Vilcea']
graph['Timisoara'] = ['Arad', 'Lugoj']
graph['Lugoj'] = ['Timisoara', 'Mehadia']
graph['Mehadia'] = ['Lugoj', 'Drobeta']
graph['Drobeta'] = ['Mehadia', 'Craiova']
graph['Craiova'] = ['Drobeta', 'Rimnicu Vilcea', 'Pitesti']
graph['Rimnicu Vilcea'] = ['Sibiu', 'Craiova', 'Pitesti']
graph['Fagaras'] = ['Sibiu', 'Bucharest']
graph['Pitesti'] = ['Rimnicu Vilcea', 'Craiova', 'Bucharest']
graph['Bucharest'] = ['Fagaras', 'Pitesti', 'Giurgiu', 'Urziceni']
graph['Giurgiu'] = ['Bucharest']
graph['Urziceni'] = ['Bucharest', 'Hirsova', 'Vaslui']
graph['Hirsova'] = ['Urziceni', 'Eforie']
graph['Eforie'] = ['Hirsova']
graph['Vaslui'] = ['Urziceni', 'Iasi']
graph['Iasi'] = ['Vaslui', 'Neamt']
graph['Neamt'] = ['Iasi']

# Defining the DFS algorithm
def dfs(graph, start, end, visited=None, path=None):
    if visited is None:
        visited = set()
    if path is None:
        path = []
    visited.add(start)
    path.append(start)

    if start == end:
        return path

    for neighbor in graph[start]:
        if neighbor not in visited:
            new_path = dfs(graph, neighbor, end, visited, path)
            if new_path:
                return new_path
    return None

# Defining the BFS algorithm
def bfs(graph, start, end):
    visited = []
    queue = [[start]]

    if start == end:
        return [start]

    while queue:
        path = queue.pop(0)
        node = path[-1]
        if node not in visited:
            neighbors = graph[node]
            for neighbor in neighbors:
                new_path = list(path)
                new_path.append(neighbor)
                queue.append(new_path)
                if neighbor == end:
                    return new_path

            visited.append(node)
    return None

# Testing the algorithms with source and destination nodes
source = 'Arad'
destination = 'Bucharest'

print("\nShortest Path From Arad To Bucharest using DFS algorithm:")
print(dfs(graph, source, destination))

print("\nShortest Path From Arad To Bucharest using BFS algorithm:")
print(bfs(graph, source, destination))
```
